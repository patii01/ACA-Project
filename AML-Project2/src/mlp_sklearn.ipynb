```python
y_trainv = train_df['class']
X_trainv = train_df.drop(columns=['class', 'filename', 'number'], axis=1)
```

```python
X_trainv = X_trainv / 255
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_trainv, y_trainv, train_size=0.8, random_state=42, stratify=y_trainv)
```

```python
from sklearn.decomposition import PCA

pca_test = PCA(.85)
pca_test.fit(X_trainv)
print(pca_test.n_components_)
```

```python
pca_test = PCA(.90)
pca_test.fit(X_trainv)
print(pca_test.n_components_)
```

```python
pca_test = PCA(.95)
pca_test.fit(X_trainv)
print(pca_test.n_components_)
```

```python
def pca(data, n_components): 

    ppca = PCA(n_components)    
    principalComponents = ppca.fit_transform(data) 

    principalDf = pd.DataFrame(data = principalComponents, columns = ['Principal Component ' + str(i+1) for i in range(principalComponents.shape[1])])

    return principalDf
```

```python
pca_train = pca(X_train, 725)

display(pca_train)
```

```python
pca_test = pca(X_test, 725)
```

### MLP sklearn

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import plot_confusion_matrix, confusion_matrix
from sklearn.metrics import classification_report

mlp = MLPClassifier(activation="relu", solver="adam", max_iter=100000, shuffle=False)
mlp.fit(X_train, y_train)

predict_test = mlp.predict(X_test)

print(confusion_matrix(y_test, predict_test))

print(classification_report(y_test, predict_test))
```

```python
kaggle_df = test_df.drop(columns=['filename', 'number'], axis=1)
```

```python
kaggle_df = kaggle_df / 255

pca_kaggle = pca(kaggle_df, 725)

predict_kaggle = mlp.predict(kaggle_df)
```

```python
submission = pd.DataFrame({'number': test_df['number'], 'class': predict_kaggle})
print(submission)
submission.to_csv('submission_mlp_PCA.csv', index=False)
```
